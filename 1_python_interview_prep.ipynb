```python
# Check the Prime number

#notes:
#Start with flag = 1 → Assume number is prime.
#If number is 0 or 1 → Directly print not_prime.
#Else loop from 2 to n-1:
#If divisible (n % i == 0) → Set flag = 0 and break.
#After loop:
#If flag == 0 → Not prime.
#Else → Prime.


def prime(n):
    flag = 1
    if n == 0 or n == 1:
        print('not_prime')
    elif n > 1:
        for i in range(2, n):
            if n % i == 0:
                flag = 0
                break
    if flag == 0:
        print("not_prime")
    else:
        print("prime")

prime(20)


```

```python
#2      Find factorial of a number
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
    fact(5)
```

```python
#10_count the words given sentence
s="sathya is the best and best is big data"
s.split()
dic={}
for i in s.split():
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1

dic
```

```python
#remove the duplicate words in the given sentence

s="sathya is the best and best is big data"
def dup(s):
    for i in s.split():
        if s.count(i)>1:
            s=s.replace(i,'')
    return s
dup(s)

```
